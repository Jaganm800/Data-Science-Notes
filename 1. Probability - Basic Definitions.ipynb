#### 1. What is an experiment?

In probability theory, an experiment refers to any operation that leads to uncertain outcomes. It could be tossing a coin, rolling a die, conducting a survey, or any other activity where the result is not certain beforehand. The main characteristics of an experiment in probability are:

__Randomness:__ The outcome of the experiment cannot be predicted with certainty. There is an element of chance involved.

__Sample Space:__ This is the set of all possible outcomes of the experiment. For example, if you're rolling a six-sided die, the sample space would be {1, 2, 3, 4, 5, 6}.

__Events:__ These are subsets of the sample space. An event is a collection of possible outcomes of an experiment. For instance, if you roll a die, the event "*getting an even number*" corresponds to the set {2, 4, 6}.

__Probability:__ This measures the likelihood of different outcomes or events occurring. It's a numerical value between 0 and 1, where 0 indicates impossibility and 1 indicates certainty.

By analyzing the sample space and assigning probabilities to different outcomes or events, we can make predictions about the likelihood of certain outcomes occurring and perform calculations to understand the behavior of random phenomena.

#### 2. Types of experiments in Probability

There are two main types of experiments in probability:

__Probabilistic or Random experiments:__  These are experiments where each outcome has an equal chance of happening or an experiment, when repeated under identical conditions donot produce the same outcomes every time.

Examples: Flipping a fair coin, rolling a balanced die, and picking a marble from a bag without looking are all examples of random experiments.

__Deterministic experiments:__ In contrast, deterministic experiments always have the same outcome under the same conditions. These are the experiments which when repeated under identical condition produced the same results or outcome are known as deterministic experiment.

For instance, measuring the length of a table with a ruler is a deterministic experiment, because if you repeat the experiment carefully, you will always get the same answer.

Experiments in probability are used to study the likelihood of events happening. By performing the experiment multiple times, we can estimate the probability of a particular event occurring. This is called experimental probability. There's also theoretical probability, which uses math to figure out the probability without needing an experiment.

#### 4. What is an outcome?

In probability theory, an outcome refers to a possible result of an experiment or a random event. It is a single element of the sample space, which is the set containing all possible outcomes of the experiment.

For example, if you roll a six-sided die, __the sample space consists of the numbers {1, 2, 3, 4, 5, 6}__, and each of these numbers represents a potential outcome of the experiment. So, if you roll the die and it lands on 3, the outcome of that particular trial of the experiment is 3.

Similarly, if you flip a coin, __the sample space consists of two outcomes: heads or tails__. Each time you flip the coin, the outcome will be either heads or tails.

In summary, an outcome is a specific result that can occur as a consequence of performing an experiment or observing a random event.

#### 5. What is a samplespace?

The sample space, often denoted by the symbol ___S___, is the set of all possible outcomes of an experiment or a random phenomenon. It encompasses every conceivable result that could occur from the experiment, including both favorable and unfavorable outcomes.

For instance:

- When rolling a six-sided die, ___the sample space is {1, 2, 3, 4, 5, 6}___, because these are all the possible outcomes.
- In a coin toss, ___the sample space is {heads, tails}___, as these are the only two possible outcomes.

- When rolling two dice at a time, the sample space will have (2*number of faces of the dice) outcomes
- When rolling $N$ dice at a time, the sample space will have ($N$*number of faces of the dice) outcomes

- When rolling two coins at a time, the sample space will have (2*number of faces of the coin) outcomes
- When rolling $N$ coins at a time, the sample space will have ($N$*number of faces of the coin) outcome

- When a dice and a coin is tossed, the sample space will have (faces of dice * faces of coin) outcomes #Applicable only to independent events

#### 6. What is an event?

In probability theory, an event is a subset of the sample space of an experiment. It represents a collection of possible outcomes of the experiment.

Formally, an event 
___E___ is any subset of the sample space ___S___. If an outcome of the experiment is in the event ___E___, we say that the event ___E___ has occurred.

Here are a few examples to illustrate:

__Rolling a die:__ Suppose you roll a six-sided die. The sample space is ___S___ = {1,2,3,4,5,6}. Now, let's define some events:

___Event E1___: *Getting an even number*. This event corresponds to the subset ___E1___={2,4,6}.

___Event E2___: *Getting a number less than 4*. This event corresponds to the subset ___E2___={1,2,3}.

___Event E3___: *Getting a number greater than 6*. Since there are no outcomes greater than 6 in the sample space, *this event is an empty set ___E3___ ={ }.*

 
 
__Flipping a coin:__ Suppose you flip a fair coin. The sample space is ___S__={heads, tails}. Now, let's define some events:

___Event F1___: *Getting heads*. This event corresponds to the subset ___F1___ ={heads}.

___Event F2___: *Getting tails*. This event corresponds to the subset ___F2___ ={tails}.

___Event F3___: *Getting heads or tails*. This event corresponds to the entire sample space ___F3___ ={heads, tails}.

#### 7. How many events can be formed from a sample space?

The number of events that can be formed from a sample space depends on the size of the sample space and the rules of set theory.

If the sample space has *n* outcomes, then the number of possible events that can be formed from it is 2***n*. This is because each subset of the sample space, including the empty set and the sample space itself, can be considered an event.

For example:

- If you roll a fair six-sided die, the sample space is {1,2,3,4,5,6}, which has n=6 outcomes. So, the number of possible events is 2**6 = 64
- If you flip a fair coin, the sample space is {"heads","tails"}, which has *n* = 2 outcomes. So, the number of possible events is 2**2 =4.

Keep in mind that this count includes the empty set and the sample space itself, which are technically considered events. Therefore, the number of non-trivial events (events other than the empty set and the sample space) is (2***n*)-2

### Intersection of sets

The intersection of two sets, denoted by $A  B$, is a fundamental concept in set theory. It refers to the elements that are common to both sets $A$ and $B$. In other words, it's the set containing all the elements that are present in both sets simultaneously.

For example, let's say you have two sets:

$A$ = {1,2,3,4}

$B$ = {3,4,5,6}

The intersection of sets $A$ and $B$ would be $AB$ = {3,4}, because these are the elements that exist in both sets $A and $B.

In set notation, you can define the intersection as:

$A  $B = {$x:xA$ and $xB$}

This means that the intersection contains all elements $x$ such that $x$ is both in set $A$ and in set $B$. If there are no common elements between the two sets, then the intersection would be an empty set, denoted by $.

### Union of sets

The union of sets is a fundamental concept in set theory, where it refers to combining all the elements from two or more sets to form a new set. In other words, the union of two sets $A$ and $B$, denoted as $AB$, contains all the elements that are in either set 
$A$ or set $B$, or both.

Mathematically, the union of sets A and B is represented as:

$AB$ = {$x:xA$ or $xB$}

This notation means that $AB$ consists of all elements $x$ such that $x$ belongs to set $A$ or $x$ belongs to set $B$, or both.

For example, let's say we have two sets: 

$A$ = {1,2,3}

$B$ = {3,4,5}

The union of sets A and B would be:

$AB$ = {1,2,3,4,5}

Here, the union contains all the elements from both sets, without any repetitions. If an element appears in both sets, it is included only once in the union.

# Mutually Exclusive Sets or Disjoint Sets

Mutually exclusive events, also known as disjoint events, are events that cannot occur simultaneously. In other words, if one event happens, the other event cannot occur at the same time. Mathematically, two events $A$ and $B$ are mutually exclusive if their intersection is an empty set, meaning they have no common outcomes.

Formally, if $A$ and $B$ are mutually exclusive events, then:

$AB = $

This means that the probability of both events occurring together is zero.

For example, consider rolling a six-sided die. Let event $A$ be rolling an odd number ({1,3,5}) and event $B$ be rolling an even number ({2,4,6}). Since no number can be both odd and even simultaneously, events $A$ and $B$ are mutually exclusive.

Another example could be tossing a coin. Let event $A$ be getting heads and event $B$ be getting tails. Since a coin cannot land with both heads and tails up at the same time, these events are also mutually exclusive.

In summary, mutually exclusive events are those that cannot happen at the same time; if one event occurs, the other event cannot occur.

# Non Mutually Exclusive Events or Joint Events

Non-mutually exclusive events, also known as joint events, are events that can occur simultaneously. In contrast to mutually exclusive events, where the occurrence of one event precludes the occurrence of the other, non-mutually exclusive events can happen together.

Mathematically, two events $A$ and $B$ are non-mutually exclusive if their intersection is not an empty set, meaning they have some common outcomes.

Formally, if $A$ and $B$ are non-mutually exclusive events, then:
    


This indicates that there are outcomes that belong to both events $A$ and $B$.

For example, consider drawing a card from a standard deck of 52 playing cards. Let event $A$ be drawing a red card and event $B$ be drawing a face card (jack, queen, or king). These events are non-mutually exclusive because there are cards in the deck that are both red and face cards (e.g., the red jack of hearts, the red queen of diamonds, etc.).

Another example could be selecting a student from a classroom based on gender and whether they participate in sports. The events "being male" and "participating in sports" are not mutually exclusive because there can be male students who participate in sports.

In summary, non-mutually exclusive events are those for which the occurrence of one event does not prevent the occurrence of the other, and they can happen together.

# Exhaustive sets
Exhaustive sets or events refer to a collection of outcomes in a probability experiment that includes all possible outcomes. In other words, when you have a set of events that cover every possible outcome that could occur in a given situation, that set is considered exhaustive.

For example, if you're rolling a six-sided die, the set of outcomes {1, 2, 3, 4, 5, 6} is exhaustive because it includes every possible result of the die roll. Another example could be when you're flipping a coin, and the set of outcomes {Heads, Tails} is exhaustive because those are the only two possibilities.

In probability theory, when you have exhaustive sets or events, it ensures that every possible outcome is considered, allowing for a comprehensive analysis of the probabilities associated with different events.

# Independent Events

Independent events are events where the occurrence of one event does not affect the probability of the other event happening. In other words, the outcome of one event has no influence on the outcome of the other event.

Mathematically, two events $$A and $B$ are independent if and only if the probability of both events occurring is the product of the probabilities of each event occurring separately. This can be expressed as:

$P(AB)=P(A)P(B)$

In words, the probability of both events $A$ and $B$ occurring is equal to the probability of event $A$ occurring multiplied by the probability of event $B$ occurring.

For example, consider flipping a fair coin twice. Let event $A$ be getting heads on the first flip and event $B$ be getting heads on the second flip. Since the outcome of the first flip does not affect the outcome of the second flip, these events are independent. The probability of getting heads on the first flip is 
$1/2$ and the probability of getting heads on the second flip is also $1/2$. Therefore, the probability of both events occurring is 
$1/21/2=1/4$.

Similarly, drawing a card from a deck of cards twice (without replacement) can also represent independent events if the probabilities of drawing certain cards are not affected by the previous draws.

In summary, independent events are those where the occurrence of one event does not influence the occurrence of the other event, and their probabilities can be multiplied to find the probability of both events happening together.
